```python
#drop columns
data1 = data.drop(['No_of_cases_min','No_of_cases_max','No_of_deaths_min','No_of_deaths_max'],axis=1)
data1.head()
```

```python
#display data year 2010
data2 = data1[(data1.Year == 2010)]
data2.head()
```

```python
data2.shape
```

```python
#2010 number of cases recorded based on continental
from matplotlib import pyplot as plt

Count1 = data2['Countinental']
NoCases = data2['No_of_cases_median']



# Figure Size
fig = plt.figure(figsize =(10, 7))
# Horizontal Bar Plot
plt.bar(Count1[0:20], NoCases[0:20])

plt.title('2010 Number of Cases based continental',fontsize=24)

plt.xlabel('Continental', fontweight ='bold', fontsize = 15)
plt.ylabel('No of cases (Million)', fontweight ='bold', fontsize = 15)

# Show Plot
plt.show()
```

```python
data3 = data1[(data1.Year == 2011)]
```

```python
#2011 number of cases recorded based on continental
from matplotlib import pyplot as plt

Count1 = data3['Countinental']
NoCases = data3['No_of_cases_median']

# Figure Size
fig = plt.figure(figsize =(10, 7))
# Horizontal Bar Plot
plt.bar(Count1[0:20], NoCases[0:20])

plt.title('2011 Number of Cases based continental',fontsize=24)

plt.xlabel('Continental', fontweight ='bold', fontsize = 15)
plt.ylabel('No of cases (Million)', fontweight ='bold', fontsize = 15)

# Show Plot
plt.show()
```

```python
#Grouping the data based on year and continental
gkk = data1.groupby(['Countinental', 'Year'])

# Print the first value in each group
gkk.first()
```

```python
#2017 Number of Death based continental
data4 = data1[(data1.Year == 2017)]
from matplotlib import pyplot as plt

Count1 = data4['Countinental']
NoDeath = data4['No_of_deaths_median']

# Figure Size
fig = plt.figure(figsize =(10, 7))
# Horizontal Bar Plot
plt.bar(Count1[0:20], NoDeath[0:20])

plt.title('2017 Number of Death based continental',fontsize=24)

plt.xlabel('Continental', fontweight ='bold', fontsize = 15)
plt.ylabel('No of death (Thousand)', fontweight ='bold', fontsize = 15)

# Show Plot
plt.show()
```
